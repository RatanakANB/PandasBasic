```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

song = pd.read_csv("../DB/spoify2019.csv")
df = song[["Rank", "Artist", "Track Name", "Artist_popularity", "Streams","tempo" , "danceability", "energy"]]

print(df.head())

pop_song = df[df["Artist_popularity"] > 90]
rank = pop_song["Artist"]. value_counts()
print('\n <Number of popular songs>')
print(rank)

x = pop_song["tempo"]
y = pop_song["Rank"]
plt.scatter(x, y, color = "red", alpha = 0.5)
plt.show
cor = pop_song.corr(method = 'pearson', numeric_only = True)
plt.matshow(cor)
plt.colorbar
plt.figure(figsize = (8, 6))
sns.heatmap(cor, annot = True)
plt.show
bts = pop_song[pop_song["Artist"] == "BTS"]
print('\n <Track list for your favorite artist>')
print(bts)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('../DB/spoify2019.csv')

# Define a tempo threshold for lively music
tempo_threshold = "Marshmello"

# Filter the dataset for lively music
lively_music_df = df[df['Artist'] == tempo_threshold]

# Save the filtered DataFrame to an Excel file
lively_music_df.to_excel(excel_writer='../DB/lively_music_playlist.xlsx', index=False)


```
